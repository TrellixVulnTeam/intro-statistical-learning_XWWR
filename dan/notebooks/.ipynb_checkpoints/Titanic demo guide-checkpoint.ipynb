### Variable Notes

**pclass**: A proxy for socio-economic status (SES)

- 1st = Upper
- 2nd = Middle
- 3rd = Lower


**age**: Age is fractional if less than 1. If the age is estimated, is it in the form of xx.5


**sibsp**: The dataset defines family relations in this way...

- Sibling = brother, sister, stepbrother, stepsister
- Spouse = husband, wife (mistresses and fiancés were ignored)


**parch**: The dataset defines family relations in this way...

- Parent = mother, father
- Child = daughter, son, stepdaughter, stepson
- Some children travelled only with a nanny, therefore parch=0 for them.


## Let's see some general patterns

We can see what the general patterns or correlations are between the features and our survival target feature.

We must always take into account that this is not perfect and that **it doesn't apply** to some features. If we have features that have a number type but this number encodes a property of the data that is not numeric, then it will be irrelevant (e.g. if we're encoding colors as 1 : yellow, 2 : green, 3: red... then plotting these numbers against other variables is plain nonsense)

So let's just recap on the features and talk a little bit about their types:

**Types**:

- **Categorical**: A feature has different categories.
- **Ordinal**: A feature has different categories that can be ordered logically.
- **Numerical**: A feature has numerical values which respect both order and proportion.

---

- survival : Categorical
- pclass : Ordinal
- name : ? (string)

- Age: Numeric

## Let's start asking the right questions!

So... we've all been thinking it right?

"Women and children first!" - The classic thing that's shouted whenever there's a shipwreck... (or we're watching a movie about a shipwreck, given that they are not that common anymore..)

**How many women and how many men survived?**

Let's start simple: How many people survived...

```python
data.Survived.value_counts()
```

```python
plt.bar(['Died','Survived'],data.Survived.value_counts())
```

Ouch.. the majority died...

How many men vs women survived?

```python
#loc -> specify columns with strings
#iloc -> specify columns with numbers

men_survived = (data.loc[:,'Sex'] == 'male') & (data.loc[:,'Survived'] == 1)
men_died = (data.loc[:,'Sex'] == 'male') & (data.loc[:,'Survived'] == 0)
```

```python
#plt.bar([list of x values], [list of y values]) #Activity: read the documentation and find out how to put a title on the graph
plt.bar(['Died', 'Survived'],[men_died.sum(), men_survived.sum()])
```

```python
women_survived = (data.loc[:,'Sex'] == 'female') & (data.loc[:,'Survived'] == 1)
women_died = (data.loc[:,'Sex'] == 'female') & (data.loc[:,'Survived'] == 0)
```

```python
plt.bar(['Died', 'Survived'],[women_died.sum(), women_survived.sum()])
```

### First pattern: Women definitely survived more often!

Let's graph it nicely!

```python
# Totally ripping off example code from the matplotlib documentation.
# That's why some variables won't make sense (like men_means, etc), I didn't bother to change them.
# (obviously it would be good to change them for clarity... but time is scarce.)

labels = ['Died', 'Survived']
men_means = [men_died.sum(), men_survived.sum()]
women_means = [women_died.sum(), women_survived.sum()]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Count')
ax.set_title('Survival by sex')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

### Let's explore age!

So let's look at age and survivability. We kind of expect a trend of younger people surviving more, let's see if this holds up.

First, let's see the distribution for all ages. Since it's a numeric variable, we'll plot some plots where we can see the distribution, such as histograms or violin plots. Box and whisker diagrams are also good but the distribution might not be as clear.

```python
#Simple: Just use pandas built in things!
data.Age.plot.hist()
```

```python
#For some reason the violinplots were having problems with the NaN (Not a Number) values present in the
# Age feature of some rows. I solved this in a hacky way by passing the data as a numpy array.
# There must be a cleaner way for this (like cleaning the data or explicitly ignorning the NaN fields.)

sns.violinplot(data=datanp[:,4])
```

```python
#Histogram using pure matplotlib and specifying 20 bins

plt.hist(datanp[:,4], bins=20)
plt.show()
```

Let's look at the age distributions of the population that survived and the one that died.

```python
pass_survived = (data.loc[:, 'Survived'] == 1)
pass_died = (data.loc[:, 'Survived'] == 0)

np_pass_survived = np.array(data[pass_survived])
np_pass_died = np.array(data[pass_died])
```

```python
sns.violinplot(data=np_pass_survived[:,4])
```

```python
sns.violinplot(data=np_pass_died[:,4])
```

```python
# Managed to plot both together, just passed a list with the two sets of data (notice the square brackets!)
# sns.violinplot(data=[set_1, set_2])


sns.violinplot(data=[np_pass_died[:,4], np_pass_survived[:,4]])
```

```python
data[pass_survived].Age.describe()
```

```python
data[pass_died].Age.describe()
```

We can see that the distributions are indeed different. This makes age a pretty good candidate for a feature we should include in an ML model to predict survival because there would be information encoded in a certain way in these distributions that would make it.
