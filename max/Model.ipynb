### Dealing with NaNs

```python
data.isnull().sum()
```

```python
data = data.dropna(subset=["Embarked"])

data.isnull().sum()
```

```python
# impute mean of Age for missing ages
mean = data['Age'].mean()

data.loc[:, ('Age')].fillna(mean, inplace=True) 
```

### Dealing with string-variables

```python
# converting type of columns to 'category'
data['Sex'] = data['Sex'].astype('category')
# Assigning numerical values and storing in another column
data['Sex_cat'] = data['Sex'].cat.codes

data['Embarked'] = data['Embarked'].astype('category')
data['Embarked_cat'] = data['Embarked'].cat.codes
data
```

```python
features = ['Pclass', 'Sex_cat', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked_cat']
X = data[features]
y = data['Survived']

X.describe()
```

```python
y.describe()
```

###  Splitting and making predictions on test data

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error

# split data into training and validation data, for both features and target
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

# Define model
dec_t_model = RandomForestClassifier(n_estimators=100, max_depth=9)

# Fit model
dec_t_model.fit(train_X, train_y)

# get predictions on validation data
val_predictions = dec_t_model.predict(val_X)

# measure of how good/bad the predictions are
print(mean_absolute_error(val_y, val_predictions))
```
